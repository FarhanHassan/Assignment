## ***13. Find Total Number of Persons Having Age Between 17 To 48 (Inclusive) Using Between Method And Also Without Using Between Method***

```python
# Using between method
count_between = df['age'].between(17, 48).sum()
# Without using between method
count_without_between = ((df['age'] >= 17) & (df['age'] <= 48)).sum()
print(count_between, count_without_between)
```

## ***17. Replace income value [<=50k with 0 and >50k with 1]***

```python
df['income'] = df['income'].apply(lambda x: 0 if x == '<=50K' else 1)
```

```python
df.head(10)
```

## ***18. Which Workclass getting The Highest Salary***

```python
highest_salary_workclass = df.groupby('workclass')['income'].mean().idxmax()
print(highest_salary_workclass)
```

## ***19. Who Has Better Chance To Get Salary >50k Male or Female***

```python
male_prop = (df[df['gender'] == 'Male']['income'] == 1).mean()
female_prop = (df[df['gender'] == 'Female']['income'] == 1).mean()

print("Male proportion:", male_prop, "Female proportion:", female_prop)
```
